```python
def calculate_evaluation_metrics(total_TP, total_pred, total_gt):
    precision = total_TP / total_pred if total_pred > 0 else 0.0
    recall = total_TP / total_gt if total_gt > 0 else 0.0
    fscore = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0.0

    return precision, recall, fscore
```

```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
from doctr.models import detection_predictor
from doctr.io import DocumentFile
from doctr.utils.geometry import detach_scores
from shapely.geometry import Polygon

def _to_absolute(geom, img_shape: tuple[int, int]) -> list[list[int]]:
    """
    Convert relative coordinates to absolute pixel values.
    If the provided geometry is given as two points ([xmin, ymin], [xmax, ymax]),
    it returns the corresponding four-corner box.
    Otherwise, it converts each point in the polygon.
    """
    h, w = img_shape
    if len(geom) == 2:
        (xmin, ymin), (xmax, ymax) = geom
        xmin, xmax = int(round(w * xmin)), int(round(w * xmax))
        ymin, ymax = int(round(h * ymin)), int(round(h * ymax))
        return [[xmin, ymin], [xmax, ymin], [xmax, ymax], [xmin, ymax]]
    else:
        return [[int(point[0] * w), int(point[1] * h)] for point in geom]

def load_gt_annotation(gt_file_path: str) -> list[tuple[list[list[float]], str]]:
    """
    Load ground-truth data from a text file.
    Each line is expected to have at least 9 comma-separated tokens:
    the first 8 tokens are the coordinates (x1,y1,x2,y2,x3,y3,x4,y4)
    and the rest is the transcription.
    Returns a list of tuples: (polygon, transcription).
    """
    gt_data = []
    if not os.path.exists(gt_file_path):
        print(f"[WARNING] GT file not found: {gt_file_path}")
        return gt_data
    with open(gt_file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            line = line.strip()
            if not line:
                continue
            tokens = line.split(',')
            if len(tokens) < 9:
                print(f"[WARNING] Not enough tokens in line: {line}")
                continue
            try:
                coords_tokens = tokens[:8]
                coords = list(map(float, coords_tokens))
                polygon = [[coords[i], coords[i+1]] for i in range(0, 8, 2)]
                transcription = ','.join(tokens[8:]).strip()
                gt_data.append((polygon, transcription))
            except Exception as e:
                print(f"Error parsing line: {line}\n{e}")
    return gt_data

def polygon_iou(poly1: list[list[int]], poly2: list[list[int]]) -> float:
    """
    Compute Intersection-over-Union (IoU) between two polygons using shapely.
    """
    try:
        p1 = Polygon(poly1)
        p2 = Polygon(poly2)
        if not p1.is_valid or not p2.is_valid:
            return 0.0
        inter = p1.intersection(p2).area
        union = p1.union(p2).area
        return inter / union if union != 0 else 0.0
    except Exception as e:
        return 0.0

# --- Set Paths ---
icdar_image_dir = '/content/Test/ch4_test_images/'
gt_annotation_dir = '/content/Test/Challenge4_Test_Task1_GT/'

# List image files (supports .png, .jpg, .jpeg)
all_image_files = [os.path.join(icdar_image_dir, f)
                   for f in os.listdir(icdar_image_dir)
                   if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Limit processing to the first 100 images.
image_files = all_image_files[:100]

# --- Prepare Documents ---
docs = DocumentFile.from_images(image_files)

# Create a base output directory for saving cropped word images and visualizations.
output_base = '/content/Cord_results'
os.makedirs(output_base, exist_ok=True)

# --- Define the Detection Predictor (DBNet) ---
det_predictor = detection_predictor(
    arch="db_resnet50",
    pretrained=True,
    assume_straight_pages=True,
    symmetric_pad=True,
    preserve_aspect_ratio=True,
    batch_size=1,
)
det_predictor.model.postprocessor.bin_thresh = 0.3
det_predictor.model.postprocessor.box_thresh = 0.1

# --- Run Inference ---
results = det_predictor(docs)

# --- Initialize Global Counters for Evaluation ---
total_gt = 0      # Count of "care" ground truth boxes (transcription != "###")
total_pred = 0    # Count of predictions (after discarding those overlapping with do-not-care regions)
total_TP = 0      # Count of true positive matches

iou_threshold = 0.5

# --- Process Results and Evaluate ---
for doc, res, img_path in zip(docs, results, image_files):
    image = cv2.imread(img_path)
    img_shape = (doc.shape[0], doc.shape[1])

    base_filename = os.path.splitext(os.path.basename(img_path))[0]
    output_folder = os.path.join(output_base, base_filename)
    os.makedirs(output_folder, exist_ok=True)

    # For an image "img_1.jpg", expected GT file is "gt_img_1.txt"
    gt_filename = 'gt_' + base_filename + '.txt'
    gt_path = os.path.join(gt_annotation_dir, gt_filename)
    gt_data = load_gt_annotation(gt_path)

    # Partition GT into care and do-not-care
    care_gt = [poly for (poly, trans) in gt_data if trans != "###"]
    dontcare_gt = [poly for (poly, trans) in gt_data if trans == "###"]
    total_gt += len(care_gt)

    predicted_polygons = []
    detached_coords, _ = detach_scores([res.get("words")])
    word_index = 0
    for coords in detached_coords[0]:
        if coords.shape == (4,):
            coords = coords.reshape(2, 2).tolist()
        else:
            coords = coords.tolist()
        abs_poly = _to_absolute(coords, img_shape)
        predicted_polygons.append(abs_poly)
        pts = np.array(abs_poly, dtype=np.int32).reshape((-1, 1, 2))
        cv2.polylines(image, [pts], isClosed=True, color=(255, 0, 0), thickness=2)
        x, y, w, h = cv2.boundingRect(pts)
        cropped_word = image[y:y+h, x:x+w]
        crop_save_path = os.path.join(output_folder, f'word_{word_index}.png')
        cv2.imwrite(crop_save_path, cropped_word)
        word_index += 1

    # Filter predictions: discard predictions overlapping do-not-care regions.
    filtered_predictions = []
    for pred in predicted_polygons:
        ignore = False
        for dnc in dontcare_gt:
            if polygon_iou(pred, dnc) >= iou_threshold:
                ignore = True
                break
        if not ignore:
            filtered_predictions.append(pred)
    total_pred += len(filtered_predictions)

    # Match predictions to "care" GT boxes.
    matched = [False] * len(care_gt)
    for pred in filtered_predictions:
        best_iou = 0.0
        best_idx = -1
        for idx, gt in enumerate(care_gt):
            if not matched[idx]:
                iou = polygon_iou(pred, gt)
                if iou > best_iou:
                    best_iou = iou
                    best_idx = idx
        if best_iou >= iou_threshold:
            total_TP += 1
            matched[best_idx] = True

    vis_save_path = os.path.join(output_folder, f'{base_filename}_vis.png')
    cv2.imwrite(vis_save_path, image)

# --- Compute Evaluation Metrics Using the Function ---
computed_precision, computed_recall, computed_fscore = calculate_evaluation_metrics(total_TP, total_pred, total_gt)

print("Computed Evaluation Metrics:")
print(f"  Precision: {computed_precision:.4f}")
print(f"  Recall:    {computed_recall:.4f}")
print(f"  F-score:   {computed_fscore:.4f}")

```
